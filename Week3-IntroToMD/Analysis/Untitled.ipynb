```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Plot energy minimizaiton

```python
# Define the path to your xvg file
file_path = 'potential.xvg'

# Initialize lists to store the data
time = []
potential = []

# Read the data from the xvg file
with open(file_path, 'r') as file:
    for line in file:
        # Skip comments and header lines
        if line.startswith(('@', '#')):
            continue
        # Split the line into time and potential values
        values = line.split()
        time.append(float(values[0]))
        potential.append(float(values[1]))

# Convert lists to numpy arrays for easier manipulation
time = np.array(time)
potential = np.array(potential)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(time, potential, label='Potential', color='b', linewidth=2)
plt.title('GROMACS Potential Energy')
plt.xlabel('Time (ps)')
plt.ylabel('Potential (kJ/mol)')
plt.grid()
plt.legend()
plt.savefig('potential_plot.png')  # Save the plot as a PNG file
plt.show()  # Display the plot
```

## Plot Production Simulation

```python

def read_xvg(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            if not line.startswith(('#', '@')):  # 
                data.append([float(x) for x in line.split()])
    return np.array(data)


rmsd_data = read_xvg('rmsd.xvg')
rmsd_xtal_data = read_xvg('rmsd_xtal.xvg')


time = rmsd_data[:, 0]
rmsd = rmsd_data[:, 1]

time_xtal = rmsd_xtal_data[:, 0]
rmsd_xtal = rmsd_xtal_data[:, 1]


plt.figure(figsize=(10, 6))
plt.plot(time, rmsd, label='RMSD', color='b',linewidth=0.2)
plt.plot(time_xtal, rmsd_xtal, label='RMSD Xtal', color='r',linewidth=0.2)


plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.title('RMSD vs Time')
plt.legend()
plt.grid(True)

plt.show()
```

Gyrate

```python
import matplotlib.pyplot as plt

# Define the function to read the xvg file
def read_xvg(file):
    time = []
    gyrate = []
    
    with open(file, 'r') as f:
        for line in f:
            # Skip comments and labels
            if line.startswith('#') or line.startswith('@'):
                continue
            # Split the data lines into time and radius of gyration
            columns = line.split()
            time.append(float(columns[0]))
            gyrate.append(float(columns[1]))
    
    return time, gyrate

# Read the gyrate.xvg file
time, gyrate = read_xvg('gyrate.xvg')

time = [i*0.001 for i in time]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, gyrate, label='Radius of Gyration')
plt.xlabel('Time (ns)')
plt.ylabel('Radius of Gyration (nm)')
plt.title('Radius of Gyration over Time')
plt.legend()
plt.grid(True)
plt.show()
```
